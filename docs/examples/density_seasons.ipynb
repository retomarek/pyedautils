# Density Plot by Season

A kernel density estimate of a measured value (e.g. room temperature),
split by season. This reveals how the distribution shifts between
seasons.

The plot is **interactive** — hover to see exact values.

```python
import plotly.io as pio
pio.renderers.default = "notebook_connected"
```

```python
import pandas as pd
from importlib import resources

data_path = resources.files("pyedautils") / "data" / "flat_temp.csv"
df = pd.read_csv(data_path, sep=";")
df.head()
```

```python
from pyedautils.plots import plot_density_seasons

fig = plot_density_seasons(
    data=df,
    title="Room Temperature Density by Season",
    xlab="Temperature (°C)",
)
fig.show()
```
