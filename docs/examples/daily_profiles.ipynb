# Daily Profiles Overview

This example demonstrates how to use `plot_daily_profiles_overview` to visualize
energy consumption patterns across seasons and weekdays.

The bundled `ele_meter.csv` dataset contains electricity meter readings at 15-minute
intervals.

```python
import pandas as pd
from importlib import resources

# Load the bundled sample data
data_path = resources.files("pyedautils") / "data" / "ele_meter.csv"
df = pd.read_csv(data_path, sep=";")
df.head()
```

```python
from pyedautils.plots import plot_daily_profiles_overview

fig = plot_daily_profiles_overview(
    data=df,
    title="Electricity Meter - Daily Profiles",
    ylab="Energy [kWh]",
    confidence=95.0,
)
fig.show()
```

## Customization

You can customize colors, season labels, and weekday labels:

```python
fig = plot_daily_profiles_overview(
    data=df,
    title="Custom Colors Example",
    ylab="Energy [kWh]",
    confidence=90.0,
    colors={"median": "red", "bounds": "orange", "fill": "rgba(255,165,0,0.2)"},
)
fig.show()
```

## Decomposed Daily Profiles

The `plot_daily_profiles_decomposed` function shows the seasonal component per weekday
after detrending the time series with a rolling mean.

```python
from pyedautils.plots import plot_daily_profiles_decomposed

fig = plot_daily_profiles_decomposed(
    data=df,
    loc_time_zone="Europe/Zurich",
    title="Electricity - Decomposed Weekly Pattern",
    ylab="delta Energy [kWh]",
)
fig.show()
```
